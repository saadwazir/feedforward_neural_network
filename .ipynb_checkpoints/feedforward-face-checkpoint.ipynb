```python
train = pd.read_csv("train.csv")    # reading the csv file
train.columns = ['File-ID','label']

test = pd.read_csv("test.csv")    # reading the csv file
test.columns = ['File-ID','label']

def conv_imgs(pd):
    train_image = []
    for i in tqdm(range(pd.shape[0])):
        img = Image.open(pd['File-ID'][i])
        img = img_to_array(img)
        img = img/255
        train_image.append(img)
    X = np.array(train_image)
    return X


X_train = conv_imgs(train)
print(X_train.shape)

X_test = conv_imgs(test)
print(X_test.shape)
```

```python
y_train = keras.utils.to_categorical(train['label'], 2)
y_test = keras.utils.to_categorical(test['label'], 2)
```

```python


X_train /= 255
X_test /= 255

X_train = np.reshape(X_train, (24044, 361))
X_test = np.reshape(X_test, (6976, 361))


print ('Data loaded.')
data =  [X_train, X_test, y_train, y_test]


```

```python
import time
def model():
    start_time = time.time()
    print ('Compiling Model ... ')
    model = Sequential()
    model.add(Dense(700, input_dim=361))
    model.add(Activation('relu'))
    model.add(Dropout(0.4))
    
    model.add(Dense(500))
    model.add(Activation('relu'))
              
    model.add(Dense(250))
    model.add(Activation('relu'))
    model.add(Dropout(0.4))
    
    model.add(Dense(50))
    model.add(Activation('relu'))
    #model.add(Dropout(0.4))
  
    model.add(Dense(2))
    model.add(Activation('softmax'))

    rms = RMSprop()
    model.compile(loss='binary_crossentropy', optimizer=rms, metrics=['accuracy'])
    print ('Model compield in {0} seconds'.format(time.time() - start_time) )
    return model
```

```python

model, losses = fk.run_network(data,model(),50,32)
```
